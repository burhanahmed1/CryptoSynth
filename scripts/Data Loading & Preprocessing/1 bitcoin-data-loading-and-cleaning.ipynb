```python
# I manually checked the datatset it is a .csv file but columns are separated via ';' instead of ','. Also remember ';' within a string "" are not separators. Load the .csv accordingly in pandas df
```

```python
import pandas as pd

df_tweets = pd.read_csv(
    "/kaggle/input/bitcoin-tweets-20160101-to-20190329/tweets.csv",
    delimiter=";",
    quotechar='"',
    encoding="utf-8",
    parse_dates=["timestamp"],
    dtype={
        "id": str,
        "user": str,
        "fullname": str,
        "url": str,
        "text": str
    }
)

df_tweets.head(10)
```

```python
df_prices = pd.read_csv("/kaggle/input/bitcoin-price-2014-2023/BTC-USD.csv", parse_dates=["Date"])
df_prices.head(10)
```

```python
print(df_tweets.dtypes)
print(df_prices.dtypes)
```

```python
def align_timeframes(df_tweets, df_prices, start_date, end_date):
    df_tweets = df_tweets[(df_tweets["timestamp"] >= start_date) & (df_tweets["timestamp"] <= end_date)]
    df_prices = df_prices[(df_prices["Date"] >= start_date) & (df_prices["Date"] <= end_date)]
    df_tweets = df_tweets.sort_values(by="timestamp")
    
    return df_tweets, df_prices

df_tweets, df_prices = align_timeframes(df_tweets, df_prices, "2016-01-01", "2018-12-31")
```

```python
import re

def clean_tweet(text):
    if isinstance(text, str): 
        text = re.sub(r"http\S+|@\S+|#\S+", "", text) 
        text = re.sub(r"[^a-zA-Z0-9\s]", "", text) 
        return text.strip()
    return "" 

df_tweets["text"] = df_tweets["text"].apply(clean_tweet)
```
