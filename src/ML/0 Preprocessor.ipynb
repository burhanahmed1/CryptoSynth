```python
import pandas as pd

# Read CSV with proper handling of quoted text and timestamp parsing
df_tweets = pd.read_csv(
    "/kaggle/input/bitcoin-tweets-20160101-to-20190329/tweets.csv",
    delimiter=";",  # Specify the delimiter
    quotechar='"',  # Handle quoted text properly
    encoding="utf-8",  # Ensure proper character encoding
    parse_dates=["timestamp"]  # Convert timestamp column to datetime
)

# Display the first 30 rows to confirm correct parsing
df_tweets.head(30)
```

```python
# Load Bitcoin price data (with columns: Date, Open, High, Low, Close, Volume)
df_prices = pd.read_csv("/kaggle/input/bitcoin-price-2014-2023/BTC-USD.csv", parse_dates=["Date"])
df_prices.head(10)
```

```python
print(df_tweets.dtypes)
print(df_prices.dtypes)
```

```python
# Align start and end dates (example: 2019-01-01 to 2022-01-01)
start_date = "2016-01-01"
end_date = "2018-12-31"

df_tweets = df_tweets[(df_tweets["timestamp"] >= start_date) & (df_tweets["timestamp"] <= end_date)]
df_prices = df_prices[(df_prices["Date"] >= start_date) & (df_prices["Date"] <= end_date)]
```
